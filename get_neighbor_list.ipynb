# 동물의 숲 마을 주민 정보 크롤링
1. 동물 주민 종 얻기
2. 각 종 별로 있는 주민들의 이름/생일/성격 가져오기
3. 주민 사진 가져오기

```python
import requests
from bs4 import BeautifulSoup
# from selenium import webdriver
import pandas as pd
```

```python
site_path='https://namu.wiki/w/%EB%8F%99%EB%AC%BC%EC%9D%98%20%EC%88%B2%20%EC%8B%9C%EB%A6%AC%EC%A6%88/%EB%93%B1%EC%9E%A5%20%EC%A3%BC%EB%AF%BC%20%EC%9D%BC%EB%9E%8C'

res=requests.get(site_path)
res.status_code
bs=BeautifulSoup(res.content,'lxml',from_encoding='utf-8')

species_result=bs.select('div.toc-indent > div.toc-indent > span.toc-item > a.wiki-link-internal')
species=[]

for i in range(len(species_result)):
    species.append(species_result[i].text)

species.insert(6,'꼬마곰') #html 모양이 달라서 추가적으로 따로 넣어줌
```

```python
species_neighbor=bs.select('div.wiki-heading-content table.wiki-table tbody')[5:]
# print(species_neighbor)
```

```python
species_count=0
# total_data=[]
neighbor_per_species=[]
for neighbors in species_neighbor:
#     flag=0
    neighbor=neighbors.select('tr')[2:]
    for features in neighbor:
        neighbor_data=[species[species_count]]
        feature=features.select('td')
        
        #column의 수가 다른 경우들이 있어서 주민들 정보에만 해당하는 표일 때의 조건들을 나눠줌
        if len(feature)==15:
            if feature[14].text=='O':
                neighbor_data.append(feature[0].text)
                neighbor_data.append(feature[3].text)
                neighbor_data.append(feature[4].text)
                neighbor_data.append(feature[5].text)
                neighbor_per_species.append(neighbor_data)
                print(neighbor_data)
        elif len(feature)==10:
            if feature[9].text=='O':
                neighbor_data.append(feature[0].text)
                neighbor_data.append(feature[3].text)
                neighbor_data.append(feature[4].text)
                neighbor_data.append(feature[5].text)
                neighbor_per_species.append(neighbor_data)
                print(neighbor_data)
        elif len(feature)==14:
            if feature[13].text=='O':
                neighbor_data.append(feature[0].text)
                neighbor_data.append(feature[3].text)
                neighbor_data.append(feature[4].text)
                neighbor_data.append(feature[5].text)
                neighbor_per_species.append(neighbor_data)
                print(neighbor_data)
        elif len(feature)==12:
            if feature[11].text=='O':
                neighbor_data.append(feature[0].text)
                neighbor_data.append(feature[3].text)
                neighbor_data.append(feature[4].text)
                neighbor_data.append(feature[5].text)
                neighbor_per_species.append(neighbor_data)
                print(neighbor_data)
        else:
#             flag=1
            species_count-=1
            break
    species_count+=1
#     if flag!=1:
#         total_data.append(neighbor_per_species)
    if species_count==len(species):
        break
```

## 가져온 정보 excel로 변환하여 excel 파일로 저장

```python
data=pd.DataFrame(neighbor_per_species,columns=['종','이름','성별','생일','성격'])
data.head(10)
```
